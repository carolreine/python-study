```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={"Último (R$)":"valor_final","Var. Dia (%)":"var_dia_pct"}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dia_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['Var_pct'] + 1)
df_principal
```

Merge = Vlookup em abas diferentes

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

Drop -> tirar a coluna (neste caso a que está redundante)

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['Variacao_rs'] = (df_principal['valor_final']- df_principal['valor_inicial'])*df_principal['Qtde. Teórica']
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={"Qtde. Teórica":"qtd_teorica"}).copy()
df_principal
```

Código feito procurando no chatGPT.
Criou uma lógica linha a linha

```python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal
```

Merge -> os dados de diferentes páginas

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.drop(columns=['Nome da Empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={"Idade em anos":"idade"}).copy()
df_principal
```

```python
df_principal['Cat_idade'] = df_principal['idade'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

Tenho uma planilha com as colunas Variacao_rs e Resultado (Subiu ou Desceu). A partir dessas colunas, monto a tabela abaixo com uso de fórmulas do excel. Se eu tivesse a mesma tabela em um dataframe do pandas, como faria para chegar nos mesmos resultados?
Maior	R$ 4.762.926.995,25
Menor	-R$ 1.807.432.634,46
Media	R$ 165.190.210,48
Media de quem subiu	R$ 448.164.250,23
Media de quem desceu	-R$ 181.109.141,75

```python
# Calculando o valor máximo
maior = df_principal['Variacao_rs'].max()

# Calculando o valor mínimo
menor = df_principal['Variacao_rs'].min()

# Calculando a média
media = df_principal['Variacao_rs'].mean()

# Filtrando os dados para calcular média de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

# Filtrando os dados para calcular média de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

print("Maior\t", f"R$ {maior:,.2f}")
print("Menor\t", f"R$ {menor:,.2f}")
print("Media\t", f"R$ {media:,.2f}")
print("Media de quem subiu\t", f"R$ {media_subiu:,.2f}")
print("Media de quem desceu\t", f"R$ {media_desceu:,.2f}")
```

Novo dataframe só com valores que subiram

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

groupby().sum() = soma agrupada por segmento.

Ele cria um vetor, então usamos o reset_index

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
## df_analise_segmento['Variacao_rs'] = df_analise_segmento['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x)) #sinal nos milhares
df_analise_segmento
```

Análise da Soma do Saldo



```python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index() #agrupar e somar
df_analise_saldo
```

Gráfico de barra

x (horizontal)
y (vertical)


```python
df_analise_saldo['Variacao_rs'] = df_analise_saldo['Variacao_rs'].apply(lambda x: '{:,.2f}'.format(x)) #sinal nos milhares
fig = px.bar(df_analise_saldo, x='Resultado', y='Variacao_rs', text='Variacao_rs', title="Variação Reais por Resultado")
fig.show()
```
